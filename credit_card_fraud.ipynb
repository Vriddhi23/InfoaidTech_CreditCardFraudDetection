```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
pd.pandas.set_option("display.max_columns",None)
credit = pd.read_csv("creditcard.csv")
credit
```

```python
credit.describe()
```

```python
credit.duplicated(keep = 'first').sum()
```

```python
credit.drop_duplicates(keep = 'first',inplace=True)
credit.duplicated().sum()
```

```python
credit.isnull().sum()
```

```python
credit['Class'].value_counts()
```

```python
legit = credit[credit.Class == 0]
fraud = credit[credit.Class == 1]
fraud.shape
```

```python
legit_sample = legit.sample(n = 473)
```

```python
new_df = pd.concat([legit_sample,fraud],axis=0)
new_df.groupby('Class').mean()
```

```python
x = new_df.drop(columns="Class",axis= 1)
y = new_df["Class"]
```

```python
sns.scatterplot(x='V1', y='V2', hue='Class', data=new_df)
plt.title('Scatter Plot of V1 vs V2')
plt.show()
```

```python
x_train,x_test,y_train,y_test = train_test_split  (x,y,test_size = 0.2, stratify = y ,random_state = 2)
```

```python
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
x_train_prediction = model.predict(x_train)
accuarcy = accuracy_score(x_train_prediction,y_train)
print("Accuarcy score : ",accuarcy)
```

```python
test_data = model.predict(x_test)
test_data_ytest = accuracy_score(test_data,y_test)
print("Test Accuracy: ",test_data_ytest)
```

```python
print("\nInput new values for the features to predict the outcome:")
new_values = []
for feature in x.columns:
    value = float(input(f"Enter value for {feature}: "))
    new_values.append(value)
new_values = np.array(new_values).reshape(1, -1)
new_values = np.array(new_values).reshape(1, -1)

# Predict the outcome for the new values
predicted_outcome = model.predict(new_values)
if predicted_outcome == 1:
    print("Predicted Outcome: Fraud")    
else:
    print("Predicted Outcome: Legitimate")
```
